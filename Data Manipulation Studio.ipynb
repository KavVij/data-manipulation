## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
df1=pd.read_csv("dataset/baltimore_9-24-2016_9-30-2017.csv")
df2=pd.read_csv("dataset/boston_9-24-2016_9-30-2017.csv")
df3=pd.read_csv("dataset/new-york_9-24-2016_9-30-2017.csv")
df4=pd.read_csv("dataset/philadelphia_9-24-2016_9-30-2017.csv")
df1
df2
df3
df4
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
for col in df1.columns:
    df1_missing = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col, round(df1_missing*100)))
    
```

```python
list_to_replace=["Type","Grade","Origin District","Environment","Quality","Condition","Appearance","Storage","Crop","Trans Mode"]
df1[list_to_replace]=df1[list_to_replace].replace({'NaN':'UNKNOWN'})
df1
```

```python
# Checking the overall percentage of missing data from the data set

total_cells = np.product(df1.shape)

missing_cells = pd.isnull(df1).sum()

total_missing = missing_cells.sum()

percentage_missing = round((total_missing/total_cells), 2) *100

print(percentage_missing, "% Missing cells from the data")  
```
